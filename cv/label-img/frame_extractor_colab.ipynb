**Change the video path of cv2.VideoCapture to point to the desired video**

```python
import cv2
vidcap = cv2.VideoCapture("/content/gdrive/My Drive/CS/CV Footage (Duke 2019)/Buoys/output.avi") #change path
success,image = vidcap.read()
count = 0
while success:
    image = image[:,100:1920-100]
    cv2.imwrite("frame%d.jpg" % count, image)     # save frame as JPEG file      
    success,image = vidcap.read()
    print('Read a new frame: ', success)
    count += 1
```
